Data Imputation

Handling with missing data

Numeric data - mean, median

Categorical data - mode - most repeated

if more than 40% or 50% data is missing, have to drop that columns

fillna() - fill null values

isna() check null values

dropna() removes rows with missing data

Age column is numeric

mean or median to fill missing data

```python
print(df_tit['Age'].mean())
```

```python
df_tit['Age'].median()
```

```python
df_tit['Age']=df_tit['Age'].fillna(df_tit['Age'].mean())
```

```python
df_tit.isna().sum()
```

```python
df_tit.info()
```

```python
df_tit.Embarked.value_counts()
```

categorical data - string - fill - mode - frequently used

```python
df_tit['Embarked']=df_tit['Embarked'].fillna(df_tit['Embarked'].mode()[0])
```

```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50, 60, 70, 80, 90])

first_five = arr[:5]
print("First five elements:")
print(first_five)

alternate_from_1 = arr[1::2]
print("\nEvery alternate element starting from index 1:")
print(alternate_from_1)

reversed_arr = arr[::-1]
print("\nReversed array:")
print(reversed_arr)
```

```python
import numpy as np

# Create two 3x3 matrices with random integers between 1 and 20
matrix1 = np.random.randint(1, 21, size=(3, 3))
matrix2 = np.random.randint(1, 21, size=(3, 3))

print("Matrix 1:")
print(matrix1)

print("\nMatrix 2:")
print(matrix2)

#  sum
sum_matrix = matrix1 + matrix2
print("\nSum:")
print(sum_matrix)

#  mul
product_matrix = matrix1 * matrix2
print("\nMUL:")
print(product_matrix)
```

```python
import numpy as np

A = np.array([[1], [2], [3]])
B = np.array([4, 5, 6])

# Element-wise sum using broadcasting
result = A + B

print("Result of A + B using broadcasting:")
print(result)
```

```python
import numpy as np

scores = np.array([65, 78, 90, 55, 88, 92, 75, 80])

mean_score = np.mean(scores)
median_score = np.median(scores)
std_deviation = np.std(scores)
max_score = np.max(scores)
min_score = np.min(scores)

print("Mean:", mean_score)
print("Median:", median_score)
print("Standard Deviation:", std_deviation)
print("Maximum Score:", max_score)
print("Minimum Score:", min_score)
```
